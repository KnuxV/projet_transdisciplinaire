```python
import pandas as pd
import os


def concat_df(path):
    """
    function to read all the files in a folder, open them as dataframes and concatenate them in a single dataframe
    file must a tabulation as sep
    """
    lst_of_files = os.listdir(path)
    lst_of_df = []
    for file in lst_of_files:
        df = pd.read_csv(path+file, sep="\t", encoding="utf-8",index_col=False)
        lst_of_df.append(df)
    df_concat = pd.concat(lst_of_df)
    return df_concat

```

```python
df_merge = concat_df("data/")
df_merge.head()
```
