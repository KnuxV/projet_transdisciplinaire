```python
# Test on ca_filtered
df = pd.read_pickle("data/countries_edges/df_country_CA_full_data_filtered")
df_edge = pd.read_pickle("data/countries_edges/df_edge_CA_full_data_filtered")
```

```python

max_edge = df_edge.total_edge.max()
print(max_edge)
df_edge
```

```python
# TEST

boulder_coords = [0, 40]
polygon_map = folium.Map(location = boulder_coords, zoom_start = 2)
```

```python
for ind, row in df.iterrows():
    coord = [row.latitude, row.longitude]
    popup = row.percentage
    radius = row.total
    if radius > 100:
        folium.RegularPolygonMarker(coord, popup = popup, fill_color = '#ff0000', number_of_sides = 8, radius=radius/50).add_to(polygon_map) 
```
