```python
path = "data/clean_data/CA_full_data_filtered.pkl"

df = pd.read_pickle(path)
df.CN = df.CN.apply(lambda x: literal_eval(x))
df
```

```python
path_country = "data/countries1.csv"
df_country = pd.read_csv(path_country, sep='\t', encoding='utf-8', index_col=False)
df_country
```

```python
df_country['total'] = df_country["name"].apply(lambda x: tot[x])
df_country["percentage"] = round(df_country.total*100 / sum(tot.values()), 1)
```

```python
tot_edge = Counter()


for index, lst in df.CN.items():
    for elem in (combinations(lst, 2)):
        if len(set(elem))>1:
            res = tuple(sorted(list(elem)))
            tot_edge[res]+=1

```

```python
lst = ['China', 'India', 'Malaysia', 'Taiwan', 'India']
for elem in combinations(lst,2):
    print(elem)
```

```python
df_edge = pd.DataFrame.from_dict(tot_edge, orient="index").reset_index()
df_edge =df_edge.rename(columns={'index':'edge', 0:'count'})
df_edge["c1"] = df_edge.edge.apply(lambda x:x[0])
df_edge["c2"] = df_edge.edge.apply(lambda x:x[1])
df_edge = df_edge.merge(df_country[["latitude", "longitude"]], how ="left", left_on=df_edge.c1, right_on=df_country.name)
df_edge =df_edge.rename(columns={'latitude':'latitude_c1', 'longitude':'longitude_c1'})
df_edge = df_edge.drop(columns="key_0")
df_edge = df_edge.merge(df_country[["latitude", "longitude"]], how ="left", left_on=df_edge.c2, right_on=df_country.name)
df_edge =df_edge.rename(columns={'latitude':'latitude_c2', 'longitude':'longitude_c2'})
df_edge = df_edge.drop(columns="key_0")
total_edges = df_edge["count"].sum()
df_edge
```

```python
lst_europe_countries = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark','Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland','Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands','Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', "Switzerland", "Norway", "San Marino", 
                       "Montenegro", "United Kingdom", "Russia", "Ukraine", "Iceland"]
```

```python
condition_europe = df_edge.c1.isin(lst_europe_countries) & df_edge.c2.isin(lst_europe_countries)
df_edge["Europe"] = df_edge.c1.isin(lst_europe_countries) & df_edge.c2.isin(lst_europe_countries)
df_edge
df_edge_europe = df_edge[condition_europe].drop(columns=['Europe']).reset_index()
df_edge_europe
```

```python
df_country["Europe"]=df_country.name.isin(lst_europe_countries)
df_country.total[df_country.Europe].describe()
df_country_europe = df_country[df_country.name.isin(lst_europe_countries)].drop(columns=["Europe"]).reset_index()
df_country_europe
```
